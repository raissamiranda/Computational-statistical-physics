## Discretized dynamics of a population


* Population tends to increase in the next generation
* With an excess of individuals, resources become scarce and the population tends to decrease in the next generation
* We will consider that the maximum population supported by the
environment is 1


$y_{n+1}$ $=$ $r$ $\times$ $y_{n}$ $\times$ $(1-y_{n})$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def equacao(y0, r):
  return r * y0 * (1 - y0)
```

```python
def mapa_logistico(n, y0, r):
  yn = y0
  eixo_y = []
  eixo_y.append(y0)
    
  for i in range(1,n):
    yn = r * yn * (1 - yn)
    y_next = yn
    eixo_y.append(y_next)

  return eixo_y
```

```python
def plota_mapa(eixo_y):
  plt.plot(eixo_y)
```

## Graphs for y0 = 0.25 and varying the r value

```python
f, axs = plt.subplots(2, 3, figsize=(15,8))

values1 = mapa_logistico(50, 0.25, 0.5)
axs[0, 0].plot(values1)
axs[0, 0].set_title('y0 = 0.25 and r = 0.5')
axs[0, 0].set_xlabel('n')
axs[0, 0].set_ylabel('population')

values2 = mapa_logistico(50, 0.25, 2.5)
axs[0, 1].plot(values2)
axs[0, 1].set_title('y0 = 0.25 and r = 2.5')
axs[0, 1].set_xlabel('n')
axs[0, 1].set_ylabel('population')

values3 = mapa_logistico(50, 0.25, 3.1)
axs[0, 2].plot(values3)
axs[0, 2].set_title('y0 = 0.25 and r = 3.1')
axs[0, 2].set_xlabel('n')
axs[0, 2].set_ylabel('population')

values4 = mapa_logistico(50, 0.25, 3.5)
axs[1, 0].plot(values4)
axs[1, 0].set_title('y0 = 0.25 and r = 3.5')
axs[1, 0].set_xlabel('n')
axs[1, 0].set_ylabel('population')

values4 = mapa_logistico(50, 0.25, 3.7)
axs[1, 1].plot(values4)
axs[1, 1].set_title('y0 = 0.25 and r = 3.7')
axs[1, 1].set_xlabel('n')
axs[1, 1].set_ylabel('population')

values5 = mapa_logistico(50, 0.5, 4.0)
axs[1, 2].plot(values5)
axs[1, 2].set_title('y0 = 0.25 and r = 4.0')
axs[1, 2].set_xlabel('n')
axs[1, 2].set_ylabel('population')
```
